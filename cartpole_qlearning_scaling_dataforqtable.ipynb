```python
import gym 
import random 
import numpy as np
```

```python
env=gym.make('CartPole-v0')
states=env.observation_space.shape[0]
actions=env.action_space.n
```

```python
print(env.observation_space)
```

```python
q_table = np.random.uniform(low=0, high=1, size=([30, 30, 50, 50] + [env.action_space.n]))
q_table.shape
```

```python
#converts value into integer array of size 4 to fit into the qtable
def get_discrete_state(state):
    discrete_state = state/np.array([0.25, 0.25, 0.01, 0.1])+ np.array([15,10,1,10])
    return tuple(discrete_state.astype(np.int))
print(get_discrete_state(env.reset()))
```

```python
total_reward=0
prior_reward=0
total=0
```

```python
import time
for episode in range(60000):
  t=time.time() 
  discrete_state=get_discrete_state(env.reset())
  episode_reward=0
  if episode%1000==0:
    print('Episode:',episode)
  done =False
  while not done:
    #0.25 is epsilon
    if np.random.random()>0.25:
      action=np.argmax(q_table[discrete_state])
    else:
      action=np.random.randint(0,actions)
    
    nextstate,reward,done,info = env.step(action)
    episode_reward+=reward 
    next_discrete_state=get_discrete_state(nextstate)

    if not done: 
      max_future_q = np.max(q_table[next_discrete_state])
      current_q = q_table[discrete_state+(action,)]
      new_q=(1-0.1)*current_q+0.1*(reward+0.95*max_future_q)
      q_table[discrete_state+(action,)] = new_q
    discrete_state=next_discrete_state
  t1=time.time() 
  episode_total=t1-t
  total+=episode_total 
  total_reward+=episode_reward 
  prior_reward=episode_reward
  if episode%1000==0:
    print("time:",total/1000)
    total=0
    print("mean reward:",total_reward/1000)
    total_reward=0 
env.close()
```
